```python
import csv

# Open the CSV file
with open('/content/encoded_file_2.csv', 'r') as f:
    # Create a CSV reader
    reader = csv.reader(f)
    # Skip the header row
    next(reader)
    # Initialize the responses list
    responses = []
    # Loop through each row in the CSV file
    for row in reader:
        # Convert the row values to integers and append them to the responses list
        response = []
        print(row)
        for val in row:
            if val != '':
                response.append((val))
        responses.append(response)
print(responses)
```

```python
model_params = {
    'active': 0,
    'reflective': 0,
    'sensing': 0,
    'intuitive': 0,
    'visual': 0,
    'verbal': 0,
    'sequential': 0,
    'global': 0,
}
```

```python
import matplotlib.pyplot as plt
def score_responses(responses, model_params):
    print(responses)
    score_list=[]
    for response in responses:
        print(response)
        for question_num, resp in enumerate(response):
          if question_num in scoring_criteria['Active/Reflective']['A'] and int(resp) == 1:
              model_params['active'] += 1
          elif question_num in scoring_criteria['Active/Reflective']['A'] and int(resp) == 0:
              model_params['reflective'] += 1

          elif question_num in scoring_criteria['Sensing/Intuitive']['S'] and int(resp) == 1:
              model_params['sensing'] += 1
          elif question_num in scoring_criteria['Sensing/Intuitive']['S'] and int(resp) == 0:
              model_params['intuitive'] += 1

          elif question_num in scoring_criteria['Visual/Verbal']['V'] and int(resp) == 1:
              model_params['visual'] += 1
          elif question_num in scoring_criteria['Visual/Verbal']['V'] and int(resp) == 0:
              model_params['verbal'] += 1

          elif question_num in scoring_criteria['Sequential/Global']['S'] and int(resp) == 1:
            model_params['sequential'] += 1
          elif question_num in scoring_criteria['Sequential/Global']['S'] and int(resp) == 0:
            model_params['global'] += 1


        if model_params["active"] > model_params["reflective"] :
          typed1 = "A"
          tscore1 = model_params["active"] - model_params["reflective"]
          print(typed1 , tscore1)
        else:
          typed1 = "B"
          tscore1 = model_params["reflective"] - model_params["active"]
          print(typed1 , tscore1)

        if model_params["sensing"] > model_params["intuitive"] :
          typed2 = "A"
          tscore2 = model_params["sensing"] - model_params["intuitive"]
          print(typed2 , tscore2)
        else:
          typed2 = "B"
          tscore2 = model_params["intuitive"] - model_params["sensing"]
          print(typed2 , tscore2)

        if model_params["visual"] > model_params["verbal"] :
          typed3 = "A"
          tscore3 = model_params["visual"] - model_params["verbal"]
          print(typed3 , tscore3)
        else:
          typed3 = "B"
          tscore3 = model_params["verbal"] - model_params["visual"]
          print(typed3 , tscore3)

        if model_params["sequential"] > model_params["global"] :
          typed4 = "A"
          tscore4 = model_params["sequential"] - model_params["global"]
          print(typed4 , tscore4)
        else:
          typed4 = "B"
          tscore4 = model_params["global"] - model_params["sequential"]
          print(typed4 , tscore4)

        score_dict = {"type1": typed1, "score1": tscore1,
                      "type2": typed2, "score2": tscore2,
                      "type3": typed3, "score3": tscore3,
                      "type4": typed4, "score4": tscore4}

        score_list.append(score_dict)

        model_params = { 'active': 0, 'reflective': 0, 'sensing': 0, 'intuitive': 0, 'visual': 0, 'verbal': 0, 'sequential': 0, 'global': 0}


    return score_list







```

```python
fin_list = score_responses(responses, model_params)
print(fin_list)
```
